复杂度分析是一种衡量算法的渐进分析法，当次要部分可以忽略时（算法的某一个部分、常数项的系数等），只考虑主要的影响因素。

多项式的写法如：$y = ax^3 + bx^2 +cx+d$
当一个算法可以在多项式的复杂度内求得结果的话，计算机大体还是能解决的，但超过多项式的复杂度，则进入到了NP-complete、NP-hard等领域，一般只能用近似算法解决。
常见的复杂度的关系如下： O(1)<O(logn)<O(n)<O(nlogn) <O(n2)<O(n3)（k次阶）<O(2^n)<O(n!)<O(n^n)。
**由于经常听到指数增长，可能会下意识得觉得阶乘比指数增长要慢，其实不然**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# only apply for one dimension array
def factorial(array):
    size = array.size
    result = np.ones(size)
    for i in range(size):
            result[i] =np.math.factorial(array[i])
    return result
point_num = 15
x = np.arange(1,point_num)
# O(1)
constant_value = np.ones(point_num-1)
# O(log(n))
log_value = np.log(x)
# O(n)
linear_value = 2*x
# O(n^2)
quadratic_value = x**2
# O(n^3)
cubic_value = x**3
# O(2^n)
exponent_value = 2**x
# O(n!)
factorial_value = factorial(x)
# O(n^n)
max_value = x**x
plt.plot(x,constant_value,label='y=1')
plt.plot(x,log_value,label='y=log(x)')
plt.plot(x,linear_value,label='y=2*x')
plt.plot(x,quadratic_value,label='y=x^2')
plt.plot(x,cubic_value,label='y=x^3')
plt.plot(x,exponent_value,label='y=2^x')
plt.plot(x,factorial_value,label='y=x!')
plt.plot(x,max_value,label='y=x^x')
plt.yscale('log')
plt.legend()
```
